Predicción del Riesgo País de Ecuador con una Red Neuronal LSTM

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import warnings
warnings.filterwarnings('ignore')
```

Configuracion de graficos

```python
sns.set_style("darkgrid")
sns.set_palette("viridis")
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
import pandas as pd

# Ruta al archivo CSV en Drive
data_path = '/content/drive/MyDrive/Investigacion M_CV/riesgo-pas (1).csv'

# Cargar el dataset
try:
    df = pd.read_csv(data_path, parse_dates=['Período'], dayfirst=False)
    print(f"Dataset '{data_path}' cargado exitosamente.")
except FileNotFoundError:
    print(f"Error: El archivo '{data_path}' no se encontró. Asegúrate de que la ruta sea correcta.")
except Exception as e:
    print(f"Ocurrió un error al cargar el archivo: {e}")
```

```python
print("="*60)
print("Información del dataset")
print("="*60)
print(f"Dimensiones del dataset: {df.shape}")
print(f"Rango de fechas: {df['Período'].min()} a {df['Período'].max()}")
print(f"Número de días con datos: {df['Período'].nunique()}")
print("\nPrimeras 5 filas del dataset:")
print(df.head())
print("\nEstadísticas descriptivas:")
print(df['Riesgo País en Puntos Básicos'].describe())
```

Procesamiento de datos

```python
# Verificar valores nulos
if df.isnull().sum().sum() > 0:
    print(f"\nSe encontraron {df.isnull().sum().sum()} valores nulos. Se procederá a llenarlos.")
    df = df.fillna(method='ffill')  # Forward fill para series temporales

# Extraer la serie temporal
serie_temporal = df['Riesgo País en Puntos Básicos'].values.reshape(-1, 1)

# Normalización de los datos (escala 0-1)
scaler = MinMaxScaler(feature_range=(0, 1))
serie_normalizada = scaler.fit_transform(serie_temporal)

print("\n" + "="*60)
print("Preprocesamiento completado")
print("="*60)
print(f"Rango de datos originales: {serie_temporal.min():.0f} - {serie_temporal.max():.0f}")
print(f"Rango de datos normalizados: {serie_normalizada.min():.4f} - {serie_normalizada.max():.4f}")
```

Creación de secuencias para el modelo LSTM

```python
def crear_secuencias(data, look_back=60):
    """
    Crea secuencias de datos para el modelo LSTM.

    Parámetros:
    - data: array con los datos normalizados.
    - look_back: número de pasos atrás para predecir el siguiente valor.

    Retorna:
    - X: características (secuencias de entrada).
    - y: etiquetas (valor a predecir).
    """
    X, y = [], []
    for i in range(look_back, len(data)):
        X.append(data[i-look_back:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

# Definir el número de pasos de tiempo (ventana de observación)
LOOK_BACK = 60  # 60 días para predecir el siguiente

# Crear secuencias
X, y = crear_secuencias(serie_normalizada, LOOK_BACK)

# Redimensionar X para LSTM: [muestras, pasos_tiempo, características]
X = X.reshape(X.shape[0], X.shape[1], 1)

print("\n" + "="*60)
print("CREACIÓN DE SECUENCIAS")
print("="*60)
print(f"Forma de X: {X.shape}")
print(f"Forma de y: {y.shape}")
```

División en conjuntos de entrenamiento y prueba

```python
# Dividir los datos en entrenamiento (80%) y prueba (20%)
tamano_entrenamiento = int(len(X) * 0.8)
X_train, X_test = X[:tamano_entrenamiento], X[tamano_entrenamiento:]
y_train, y_test = y[:tamano_entrenamiento], y[tamano_entrenamiento:]

print("\n" + "="*60)
print("DIVISIÓN DE DATOS")
print("="*60)
print(f"Muestras de entrenamiento: {X_train.shape[0]}")
print(f"Muestras de prueba: {X_test.shape[0]}")
```

Construcción del modelo  LSTM

```python
# Inicializar el modelo
modelo_lstm = Sequential()

# Capa LSTM 1
modelo_lstm.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
modelo_lstm.add(Dropout(0.2))

# Capa LSTM 2
modelo_lstm.add(LSTM(units=50, return_sequences=False))
modelo_lstm.add(Dropout(0.2))

# Capa densa de salida
modelo_lstm.add(Dense(units=1))

# Compilar el modelo
modelo_lstm.compile(optimizer='adam', loss='mean_squared_error')

# Resumen del modelo
print("\n" + "="*60)
print("ARQUITECTURA DEL MODELO LSTM")
print("="*60)
modelo_lstm.summary()
```

Entrenamiento del modelo

```python
# Configurar early stopping para evitar sobreajuste
early_stop = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)

# Entrenar el modelo
print("\n" + "="*60)
print("ENTRENAMIENTO DEL MODELO")
print("="*60)
print("Iniciando entrenamiento...")
historia = modelo_lstm.fit(
    X_train, y_train,
    epochs=100,
    batch_size=32,
    validation_data=(X_test, y_test),
    callbacks=[early_stop],
    verbose=1
)
```

Evaluación del codigo

```python
# Realizar predicciones
y_train_pred = modelo_lstm.predict(X_train)
y_test_pred = modelo_lstm.predict(X_test)

# Invertir la normalización
y_train_pred_des = scaler.inverse_transform(y_train_pred)
y_test_pred_des = scaler.inverse_transform(y_test_pred)
y_train_des = scaler.inverse_transform(y_train.reshape(-1, 1))
y_test_des = scaler.inverse_transform(y_test.reshape(-1, 1))

# Calcular métricas de rendimiento
def calcular_metricas(y_real, y_pred):
    """
    Calcula métricas de evaluación para el modelo.
    """
    rmse = np.sqrt(mean_squared_error(y_real, y_pred))
    mae = mean_absolute_error(y_real, y_pred)
    r2 = r2_score(y_real, y_pred)
    return {'RMSE': rmse, 'MAE': mae, 'R2': r2}

metricas_train = calcular_metricas(y_train_des, y_train_pred_des)
metricas_test = calcular_metricas(y_test_des, y_test_pred_des)

print("\n" + "="*60)
print("EVALUACIÓN DEL MODELO")
print("="*60)
print("\nMétricas en el Conjunto de Entrenamiento:")
print(f"RMSE: {metricas_train['RMSE']:.2f} puntos básicos")
print(f"MAE:  {metricas_train['MAE']:.2f} puntos básicos")
print(f"R²:   {metricas_train['R2']:.4f}")

print("\nMétricas en el Conjunto de Prueba:")
print(f"RMSE: {metricas_test['RMSE']:.2f} puntos básicos")
print(f"MAE:  {metricas_test['MAE']:.2f} puntos básicos")
print(f"R²:   {metricas_test['R2']:.4f}")
```

Visualización de resultados

```python
# Gráfico 1: Pérdida del modelo durante el entrenamiento
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(historia.history['loss'], label='Pérdida Entrenamiento', color='blue')
plt.plot(historia.history['val_loss'], label='Pérdida Validación', color='orange')
plt.title('Evolución de la Pérdida del Modelo')
plt.xlabel('Épocas')
plt.ylabel('Pérdida (MSE)')
plt.legend()
plt.grid(True)

# Gráfico 2: Comparación de predicciones en el conjunto de prueba
plt.subplot(1, 2, 2)
plt.plot(y_test_des, label='Valores Reales', color='green', linewidth=2)
plt.plot(y_test_pred_des, label='Predicciones', color='red', linestyle='--', linewidth=2)
plt.title('Predicción del Riesgo País en el Conjunto de Prueba')
plt.xlabel('Muestras')
plt.ylabel('Riesgo País (Puntos Básicos)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Gráfico 3: Predicción completa (entrenamiento + prueba) vs. Datos reales
# Crear un array para almacenar todas las predicciones
predicciones_completas = np.zeros_like(serie_temporal)
predicciones_completas[LOOK_BACK:] = scaler.inverse_transform(
    modelo_lstm.predict(X).reshape(-1, 1)
)

plt.figure(figsize=(14, 7))
plt.plot(df['Período'], serie_temporal, label='Datos Reales', color='blue', linewidth=1.5)
plt.plot(df['Período'][LOOK_BACK:], predicciones_completas[LOOK_BACK:],
         label='Predicciones del Modelo', color='red', linestyle='--', linewidth=1.5)
plt.title('Predicción de la Serie Temporal del Riesgo País con LSTM')
plt.xlabel('Fecha')
plt.ylabel('Riesgo País (Puntos Básicos)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

Analisis y resultados

```python
# Calcular el error porcentual absoluto medio (MAPE) para el conjunto de prueba
def calcular_mape(y_real, y_pred):
    """
    Calcula el Error Porcentual Absoluto Medio (MAPE).
    """
    return np.mean(np.abs((y_real - y_pred) / y_real)) * 100

mape_train = calcular_mape(y_train_des.flatten(), y_train_pred_des.flatten())
mape_test = calcular_mape(y_test_des.flatten(), y_test_pred_des.flatten())

print("\n" + "="*60)
print("Análisis adicional")
print("="*60)
print(f"MAPE Entrenamiento: {mape_train:.2f}%")
print(f"MAPE Prueba: {mape_test:.2f}%")

# Identificar picos y valles en el conjunto de prueba
indice_max_real = np.argmax(y_test_des)
indice_min_real = np.argmin(y_test_des)
indice_max_pred = np.argmax(y_test_pred_des)
indice_min_pred = np.argmin(y_test_pred_des)

print(f"\nValor máximo real: {y_test_des[indice_max_real][0]:.0f} puntos básicos")
print(f"Valor máximo predicho: {y_test_pred_des[indice_max_pred][0]:.0f} puntos básicos")
print(f"Valor mínimo real: {y_test_des[indice_min_real][0]:.0f} puntos básicos")
print(f"Valor mínimo predicho: {y_test_pred_des[indice_min_pred][0]:.0f} puntos básicos")

print("\n" + "="*60)
print("Conclusiones")
print("="*60)
print("""
1. El modelo LSTM ha demostrado ser efectivo para predecir la serie temporal del Riesgo País.
2. Los valores de RMSE y MAE en el conjunto de prueba indican una precisión aceptable.
3. El coeficiente de determinación (R²) muestra que el modelo explica una parte significativa de la varianza.
4. Las visualizaciones confirman que el modelo captura adecuadamente la tendencia general y los cambios abruptos.
""")
```
